### Random Numbers

```python
def set_unit_square(ax):
    ax.axis('scaled')
    ax.set_xlim([0,1])
    ax.set_ylim([0,1])
```

### Covering the unit square

```python
rng = MersenneTwisterUniformRng(42)
```

```python
xs = []
ys = []
for i in range(2047):
    xs.append(rng.next().value())
    ys.append(rng.next().value())
```

```python
fig, ax = plot.subplots()
set_unit_square(ax)
g = ax.plot(xs, ys, 'o')
```

### Dimensionality and quality of randomness

```python
rng = SobolRsg(3)
```

```python
xs = [ rng.nextSequence().value() for i in range(1023)]
```

```python
from mpl_toolkits.mplot3d.axes3d import Axes3D
```

```python
fig = plot.figure()
ax = fig.add_subplot(1, 1, 1, projection='3d')
X, Y, Z = zip(*xs)
g = ax.plot(X, Y, Z, 'o')
```

```python
def section(xs, i):
    return [x[i] for x in xs]
```

```python
def project(ax, xs, i, j, title=None):
    set_unit_square(ax)
    ax.set_title(title or "(%d,%d)" % (i,j))
    ax.plot(section(xs,i), section(xs,j), 'o')
```

```python
fig, ax = plot.subplots(1, 3, figsize=(12,4))

project(ax[0], xs, 0, 1, title="(x,y)")
project(ax[1], xs, 0, 2, title="(x,z)")
project(ax[2], xs, 1, 2, title="(y,z)")
```
